Задание 1
Получить интуицию, на что влияют параметры распределений. Построить по 5 функций плотностей с разными параметрами для каждого распределения: нормальное, экспоненциальное, Стьюдента на одном графике.

```python
import numpy as np
import matplotlib.pyplot as plt
from math import gamma
```

```python
def gauss(x, m, sigma):
    a = 1/np.sqrt(2*np.pi*sigma)
    b = ((x-m)**2)/2*sigma
    c = np.exp(-b)
    return a*c
```

```python
xs = np.arange(-8, 8, 0.05)
p1, = plt.plot(xs, [gauss(x, 0, 0.5) for x in xs], label='$\ x,sigma = 0, 0.5$')
p2, = plt.plot(xs, [gauss(x, 0, 1) for x in xs], label='$\ x,sigma = 0, 1$')
p3, = plt.plot(xs, [gauss(x, 0, 2) for x in xs], label='$\ x,sigma = 0, 2$')
p4, = plt.plot(xs, [gauss(x, -2, 0.5) for x in xs], label ='$\ x,sigma =-1, 0.5$')
p5, = plt.plot(xs, [gauss(x, 2, 2) for x in xs], label ='$\ x,sigma =1, 2$')
plt.legend()
plt.show()
```

```python
def expon(x, lambd):
    if x < 0:
        return 0
    else:
        return lambd * np.exp(-lambd*x)
```

```python
xs = np.arange(0, 8, 0.05)
p1, = plt.plot(xs, [expon(x, 0.3) for x in xs], label='$\ lambda = 0.3$')
p2, = plt.plot(xs, [expon(x, 0.5) for x in xs], label='$\ lambda = 0.5$')
p3, = plt.plot(xs, [expon(x, 1.) for x in xs], label='$\ lambda = 1$')
p4, = plt.plot(xs, [expon(x, 1.5) for x in xs], label ='$\ lambda = 1.5$')
p5, = plt.plot(xs, [expon(x, 2) for x in xs], label ='$\ lambda = 2$')
plt.legend()
plt.show()
```

```python
def stud(x,n):
    a = gamma((n+1)/2)
    b = np.sqrt((n*np.pi))*gamma(n/2)
    c = (1+ x**2/n)**(-(n+1)/2)
    return (a/b)*c
```

```python
xs = np.arange(-8, 8, 0.05)
p1, = plt.plot(xs, [stud(x, 1) for x in xs], label='$\ n = 1$')
p2, = plt.plot(xs, [stud(x, 2) for x in xs], label='$\ n = 2$')
p3, = plt.plot(xs, [stud(x, 10) for x in xs], label='$\ n = 10$')
p4, = plt.plot(xs, [stud(x, 15) for x in xs], label ='$\ n = 15$')
p5, = plt.plot(xs, [stud(x, 341) for x in xs], label ='$\ n = 341$')
plt.legend()
plt.show()
```
