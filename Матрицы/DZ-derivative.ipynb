#### Уровень 0
посчитайте значение 1-ой и 2-ой производной фунции

f(x) = x^5 + 4sin(2x) + cos(3x+3) в точке x = 1

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from scipy.misc import derivative
```

```python
def f(x):
    return x**5 + 4* np.sin(2*x) + np.cos(3*x+3)
derivative(f,1.0, dx = 1e-6)
```

```python
derivative(f,1.0, n=2, dx = 1e-6)
```

### Уровень 1**¶
Постройте график первой и второй производной функции

(sin(2x+1))^5

на отрезке [-5; 5]

```python
def g(x):
    return (np.sin(2*x+1))**5
```

```python
x = np.arange(-5,5,0.0001)
dg1 = list()
gx = g(x)
for el in x:
    dg1.append(derivative(g, el, dx = 1e-6))
plt.plot(x,dg1)
```

```python
x = np.arange(-5,5,0.0001)
dg2 = list()
gx = g(x)
for el in x:
    dg2.append(derivative(g, el, n=2, dx = 1e-6))
plt.plot(x,dg2)
```
