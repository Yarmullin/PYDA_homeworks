```python
import pandas as pd
import numpy as np
```

```python
id_names = np.arange(1,29,1)
list_id = [1,2,3,4,5,11,13,23]
```

```python
df = pd.read_csv('horse_data.csv', names = list(id_names), na_values = '?')
df.info()
df_new = df.filter(items = list_id)
for i in list_id:
    pd.to_numeric(df_new[i])
```

```python
dict_names = {1:'surgery', 2:'Age', 3:'Hospital_Number', 4:'rectal_temperature', 
              5:'pulse', 11:'pain', 13:'abdominal_distension', 23:'outcome'}
df_new.rename(columns=dict_names, inplace=True)
```

```python
describe= df_new.describe().T.reset_index()
dict_modes = {}
for v in dict_names.values():
    dict_modes[v] = df_new[v].mode()[0]
modes = pd.DataFrame.from_dict(dict_modes, orient='index').reset_index()
modes.rename(columns = {0:'mode'}, inplace = True)
```

```python
df_new.info()
describe.merge(modes, how='left')
```

### Столбец Rectal temperature
ректальная температура 
          - линейная 
          - в градусах Цельсия.
          - Повышенная температура может возникнуть из-за инфекции. 
          - температура может снижаться, когда животное находится в позднем шоке 
          - нормальная температура составляет 37,8 
          - этот параметр обычно изменяется по мере развития проблемы, 
               например. может вначале быть нормальным, затем стать повышенным из-за 
                   поражения, переходя обратно в нормальный диапазон, когда 
                   лошадь впадает в шок 
В данном столбце отсутствует 60 значений или 20%.
Глобальных выбросов мы не видим. Медиана, как и среднее колеблятся около значения 38,2. При условии что датасет у нас о лошадях с фактом наличия колики, то можем сказать, что такие значения нормальны.
Величина размаха и СКО между собой достаточно сильно разнятся, это говорит нам о том, что в массе своей значения температуры не столь вариативны, однако, встречаются некоторые аномалии.

#### Работа с выбросами

```python
q1 = df_new.rectal_temperature.quantile(0.25)
q3 = df_new.rectal_temperature.quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = df_new[df_new.rectal_temperature.between(lower_bound, upper_bound, inclusive=True)].sort_values('rectal_temperature')
remove_outliers
```

```python
lower_outliers = df_new[df_new.rectal_temperature.between(df_new.rectal_temperature.min(), lower_bound, inclusive=True)].sort_values('rectal_temperature')
lower_outliers
```

```python
upper_outliers = df_new[df_new.rectal_temperature.between(upper_bound, df_new.rectal_temperature.max(), inclusive=True)].sort_values('rectal_temperature')
upper_outliers
```

Предлагаю не выкидывать из выборки данные выбросы, т.к. аномалий мы не видим. Вполне вероятно что лошади во время лечения могли иметь как повышенные, так и пониженные температуры.

#### Столбец Pulse
пульс 
          - линейный 
          - частота пульса в ударах в минуту 
          - отражает состояние сердца: 30-40 нормально для взрослых 
          - редко бывает ниже нормы, хотя у спортивных лошадей
            может иметь частоту 20-25 
          - животные с болезненными поражениями или страдающие от шока кровообращения 
            могут иметь повышенную частоту сердечных сокращений 
            
Отсутствует 8% значений. Есть подозрение, что возможны выбросы. Я, конечно, не ветеринар, но пульс в 184 удара в минуту при норме в 30-40 весьма подозрителен. Разница между размахом и СКО говорит нам о вариативности значений. Мода, среднее и медиана также существенно различаются друг от друга.

#### Работа с выбросами

```python
q1 = df_new.pulse.quantile(0.25)
q3 = df_new.pulse.quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = df_new[df_new.pulse.between(lower_bound, upper_bound, inclusive=True)].sort_values('pulse')
remove_outliers
```

```python
lower_outliers = df_new[df_new.pulse.between(df_new.pulse.min(), lower_bound, inclusive=True)].sort_values('pulse')
lower_outliers
```

```python
upper_outliers = df_new[df_new.pulse.between(upper_bound, df_new.pulse.max(), inclusive=True)].sort_values('pulse')
upper_outliers
```

С одной стороны, выбросы похожи на аномалии, но, если проаназировать по данным строчкам другие показатели, то данные выбросы можно интерпретировать как неудачные операции, в результате которых лошади не выжили. Предлагаю оставить.
