#### DecisionTreeRegressor

```python
#пробуем найти лучшее дерево решений
scaler = StandardScaler()
test_param = DecisionTreeRegressor(max_depth = 2)
data8 = pd.DataFrame(scaler.fit_transform(data7[['RM', 'LSTAT', 'NOX', 'DIS', 'CHAS', 'AGE', 'TAX', 'PTRATIO', 'B', 'MEDV']]), columns = ['RM', 'LSTAT', 'NOX', 'DIS', 'CHAS', 'AGE', 'TAX', 'PTRATIO', 'B', 'MEDV'])
y8 = data8['MEDV']
del data8['MEDV']
X8 = data8
X8_train, X8_test, y8_train, y8_test = train_test_split(X8, y8, test_size=0.3, random_state=42)
```

```python
%%time

param_grid = {'max_depth':list(range(1,10)),
              'criterion': ['mse', 'friedman_mse', 'mae']
}
grid = GridSearchCV(test_param, param_grid, cv=10, scoring='neg_mean_absolute_error')

grid.fit(X8_train,y8_train)
```

```python
print(grid.best_estimator_)
a,b = grid.best_params_.values()
```

```python
grid.score(X8_test, y8_test)
```

```python
grid.predict(X8_test)
```

```python
#scal_model
plt.plot(pd.DataFrame(zip(y8_test, grid.predict(X8_test)), columns = ['y_test', 'pred']))
```

```python
#применим получившиеся параметры, чтобы получить итоговую модель
model = make_pipeline(StandardScaler(), DecisionTreeRegressor(criterion = a, max_depth = b))
```

```python
data9 = pd.DataFrame((data7[['RM', 'LSTAT', 'NOX', 'DIS', 'CHAS', 'AGE', 'TAX', 'PTRATIO', 'B', 'MEDV']]), columns = ['RM', 'LSTAT', 'NOX', 'DIS', 'CHAS', 'AGE', 'TAX', 'PTRATIO', 'B', 'MEDV'])
y_fin = data9['MEDV']
del data9['MEDV']
X_fin = data9
X_fin_train, X_fin_test, y_fin_train, y_fin_test = train_test_split(X_fin, y_fin, test_size=0.3, random_state=42)
```

```python
model.fit(X_fin_train,y_fin_train)
```

```python
model.score(X_fin_test, y_fin_test)
```

```python
#итоговая модель.
plt.plot(pd.DataFrame(zip(y_fin_test, model.predict(X_fin_test)), columns = ['y_test', 'pred']))
```
