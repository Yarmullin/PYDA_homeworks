# Метрики precision и recall

Данные содержат признаки пользователей и целевую переменную affair - была ли связь на стороне

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('affair_data.csv')
data.head()
```

```python
data.info()
```

```python
from sklearn.linear_model import LogisticRegression

# импортируем метод для автоматической разбивки на обучающую и тестовую выборки
# раньше использовали from sklearn.cross_validation import train_test_split  
from sklearn.model_selection import train_test_split
```

Формируем набор признаков и целевую переменную

```python
X = data[data.columns[:-1]]
X.head(2)
```

```python
Y = data['affair']
```

```python
model = LogisticRegression()
```

Разбиваем данные на обучающую и тестовую выборки в соотношении 70 / 30

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state = 0)
```

Обучаем модель

```python
model.fit(X_train, Y_train)
```

Получаем прогнозные значения модели (переменная predictions)

```python
predictions = model.predict_proba(X_test)
```

```python
for line in zip( predictions[:, 1], Y_test ):
     print( line )
```

```python
model.score(X_test, Y_test)
```

Построим набор моделей для значений random_state от 0 до 9. Получите model.score для каждого варианта.

```python
for i in range(10):
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state = i)
    model = LogisticRegression()
    model.fit(X_train, Y_train)
    print("{}: {}".format(i, model.score(X_test, Y_test)))
```

### Оценка качества модели

Считаем accuracy последней модели

```python
tp = 0 # True positive
fp = 0 # False positive
fn = 0 # False negative
tn = 0 # True negative

predictions = model.predict_proba( X_test )
for predicted_prob, actual in zip( predictions[:, 1], Y_test ):    
    if predicted_prob >= 0.5:
        predicted = 1
    else:
        predicted = 0
    
    if predicted == 1:
        if actual == 1:
            tp += 1
        else:
            fp += 1
    
    else:
        if actual == 1:
            fn += 1
        else:
            tn += 1
```

```python
# какая точность получилась?
(tp + tn) / (tp + fp + fn + tn)
```

```python
precision = tp / (tp + fp)
precision
```

### Используем готовые библиотеки

```python
# метод для построения графика precision-recall
from sklearn.metrics import precision_recall_curve
```

```python
pr, re, thres = precision_recall_curve( Y_test, predictions[:, 1] )
```

```python
# посмотрим что получилось
for line in zip( pr, re, thres ):
    print( line )
```

```python
from sklearn.metrics import average_precision_score
```

```python
# площадь под кривой
average_precision_score( Y_test, predictions[:, 1] )
```

### ROC

```python
# посчитаем ROC-кривую (Receiver Operating Characteristic)
from sklearn.metrics import roc_curve
```

```python
# получаем значения false и true positive rate для различных значений порога
fpr, tpr, thres = roc_curve( Y_test, predictions[:, 1] )
```

```python
for line in zip( fpr, tpr, thres ):
    print( line )
```

```python
plt.plot( fpr, tpr )
```

### Ручное построение ROC-кривой

```python
n_class = len(Y_test) - sum(Y_test)
p_class = sum(Y_test)
fpr_ = 1/n_class
tpr_ = 1/p_class
rows = list(range(1, len(Y_test)+1,1))
fpr_accum = 0
```

```python
a = sorted(zip(Y_test, predictions[:,1]), key=lambda x: x[1], reverse=True)
list_fpr = []
fpr_accum = 0
list_tpr = []
tpr_accum = tpr_
i = 0
for el in a:
    if el[0] == 1:
        list_fpr.append(fpr_accum)
        tpr_accum = tpr_accum + tpr_
        list_tpr.append(tpr_accum)
    else:
        fpr_accum = fpr_accum + fpr_
        list_fpr.append(fpr_accum)
        list_tpr.append(tpr_accum)
```

```python
D_roc = pd.DataFrame(a, columns = ['Y_test', 'predictions'])
D_roc['fpr'] = list_fpr
D_roc['tpr'] = list_tpr
```

```python
plt.plot(D_roc['fpr'], D_roc['tpr'], label = 'custom')
plt.plot( fpr, tpr, label = 'auto' )
plt.legend( loc = 0 )
plt.show()
```
