#### 3. Рассчитайте и визуализировать корреляционную матрицу для количественных переменных.
Определите две самые скоррелированные и две наименее
скоррелированные переменные.

```python
df.corr()
```

```python
corr_max = df.corr().abs()
corr_max.replace([1], [0], inplace=True)
corr_max
```

```python
# ищем 2 самые скоррелированные величины, берем матрицу по модулю, т.к. возможна и обратная корреляция
indexes = pd.DataFrame(corr_max.idxmax())
indexes.rename(columns = {0:'indexes2'}, inplace=True)
values = pd.DataFrame(corr_max.max())
values.rename(columns = {0:'values'}, inplace=True)
max_corr = indexes.join(values)
max_corr.sort_values('values', ascending=False)
```

```python
# прямая
corr = df.corr()
corr.replace([1], [0], inplace=True)
indexes2 = pd.DataFrame(corr.idxmax())
indexes2.rename(columns = {0:'indexes2'}, inplace=True)
values2 = pd.DataFrame(corr.max())
values2.rename(columns = {0:'values'}, inplace=True)
positive_corr = indexes2.join(values2)
positive_corr.sort_values('values', ascending=False)
```

```python
# обратная
corr_neg = df.corr()
indexes3 = pd.DataFrame(corr_neg.idxmin())
indexes3.rename(columns = {0:'indexes2'}, inplace=True)
values3 = pd.DataFrame(corr_neg.min())
values3.rename(columns = {0:'values'}, inplace=True)
negative_corr = indexes3.join(values3)
negative_corr.sort_values('values')
```

```python
# ищем 2 слабоскоррелированные величины, берем матрицу по модулю, т.к. возможна и обратная корреляция
corr_min = df.corr().abs()
indexes1 = pd.DataFrame(corr_min.idxmin())
indexes1.rename(columns = {0:'indexes2'}, inplace=True)
values1 = pd.DataFrame(corr_min.min())
values1.rename(columns = {0:'values'}, inplace=True)
min_corr = indexes1.join(values1)
min_corr.sort_values('values')
```

2. Наименее скоррелированные величины это: стаж работы и травматичность. То есть вероятность травмы возможна как у новичков, так и у опытных сотрудников.

#### 7. Проверить гипотезу, что сотрудники с высоким окладом проводят на работе больше времени, чем сотрудники с низким окладом

Гипотеза Н0: Сотрудники с высоким и низким окладами проводят одинаковое кол-во времени на работе.

Гипотеза H1: Сотрудники с высоким и низким окладами проводят разное кол-во времени на работе.

```python
df_low = df[df.salary == 'low']
df_high = df[df.salary == 'high']
alpha = 0.05
result = st.ttest_ind(df_low['average_montly_hours'], df_high['average_montly_hours'], equal_var=False)
print(result)
if (result.pvalue < alpha):
    print('Отвергаем нулевую гипотезу, у сотрудников разное кол-во времени на работе.')
else:
    print('Не отвергаем нулевую гипотезу, сотрудники с разными окладами проводят примерно одинаковое кол-во времени на работе.')
```

● Доля сотрудников с повышением за последние 5 лет

● Средняя степень удовлетворенности

● Среднее количество проектов

```python
df_left = df[df.left == 1]
a = df_left[df_left.promotion_last_5years == 1].promotion_last_5years.sum()
b = df_left.promotion_last_5years.count()
promotion = a / b
mean_satisfaction = df_left.satisfaction_level.mean()
mean_projects = df_left.number_project.mean()
print('Показатели среди уволившихся:')
final = pd.DataFrame({'index':['share_of_promo', 'mean_satisfaction', 'mean_number_project'], 'values': [promotion, mean_satisfaction, mean_projects]})
final
```

```python
df_right = df[df.left == 0]
a = df_right[df_right.promotion_last_5years == 1].promotion_last_5years.sum()
b = df_right.promotion_last_5years.count()
promotion = a / b
mean_satisfaction = df_right.satisfaction_level.mean()
mean_projects = df_right.number_project.mean()
print('Показатели среди действующих сотрудников:')
final = pd.DataFrame({'index':['share_of_promo', 'mean_satisfaction', 'mean_number_project'], 'values': [promotion, mean_satisfaction, mean_projects]})
final
```

#### 9. Разделить данные на тестовую и обучающую выборки. Построить модель LDA, предсказывающую уволился ли сотрудник на основе имеющихся факторов (кроме department и salary). Оценить качество модели на тестовой выборке.

```python
del df['salary']
del df['department']
df
data_train, data_test = train_test_split(df, test_size=0.3)
```

```python
data_train
```

```python
data_test
```

```python
y_train, y_test = data_train['left'], data_test['left']
del data_train['left']
del data_test['left']
```

```python
lda = LinearDiscriminantAnalysis()
```

```python
lda.fit(data_train, y_train)
```

```python
lda.predict(data_test)
```

```python
result = pd.DataFrame([y_test.values, lda.predict(data_test)]).T
```

```python
result
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, lda.predict(data_test))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, lda.predict(data_test))
```

### Выводы
Модель показала не самые лучшие результаты, в частности, точность прогнозирования составила 76,42%.
Также следует выделить  матрицу неточностей: мы получили достаточно большое число ложно-отрицательных результатов (17,2%), т.е. модель определила реально уволенных сотрудников в действующие.
Что касается ложно-положительных результатов, то их значительно меньше, чем ложно-отрицательных (6,4%), но результат тоже нельзя игнорировать или списать в статистическую погрешность.
Также модель показала очень плохие результаты в определении уволенных сотрудников, все 25,3% реально уволенных сотрудников модель "угадала".
