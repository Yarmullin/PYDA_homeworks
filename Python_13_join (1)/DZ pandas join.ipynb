В задаче сквозной аналитики вам предоставили данные по местоположению пользователей. Т. е. для каждого user_id известна последовательность координат (широта/долгота), когда они требовались приложению для полноценной работы. Как бы вы добавили эти сведения в таблицу визитов и покупок? Для составления ответа можно использовать вопросы:
У каждого пользователя известен набор координат. А для связывания с визитом или фактом покупки скорее всего потребуется одно-два числа. Как их получить?
Наборы координат одного и того же пользователя могут быть значительно удалены друг от друга. Как это отразится на вопросе расчетах пункта 1?
Какие дополнительные признаки можно получить из координат? Ведь это просто числа, которые сами по себе мало что дают.

Для того, чтобы добавить данные по гео-позиции в таблицу с визитами нужна однозначная связка между таблицами, т.е. необходим какой-то ключ. В данном случае вижу 2 варианта: 
1. если в таблице с геопозицией имеется visit_id, т.е. таблица вида:visit_id|user_id|geo_coordinate1|geo_coordinate2, тогда мы просто через leftjoin приклеиваем их к таблице визитов (можно и через outer, если нам нестрашно попортить таблицу визитов непонятными визитами с пустыми полями).
2. если в таблицах есть дата-время в формате timestamp, или в принципе указано дата-время с точностью до секунды транзакции/визита, тогда создаем в обеих таблицах некий индекс: сцепка двух столбцов: timestamp & "/" & userId, после, через получившийся ключ приклеиваем к таблице визитов гео-координаты через left или outer, в зависимости от степени извращенности заказчика и исполнителя:). В данном случае обязательно нужен разделитель какой-нибудь при склейке, т.к. можем случайно наплодить дубль, в работе я обычно использую символ /.

###### Есть еще третий вариант, самый извращенный и мутный, если у нас в геокоординатах таблица вида: userId|geo1|geo2. 
Тогда нам надо анализировать таблицу и делать из нее некий свод, где у нас не будет дублей у userId, а вместо кучи гео-координат по визитам, мы выберем, например, наиболее часто встречающуюся долготу у каждого юзверя, получим таблицу вида userId|geo1. Для пользователей с равномерной структурой, например, всего 2 визита с разными координатами, примем за правило брать самую "позднюю" долготу(будем считать, что данные в нашу таблицу попадают в самую последнюю строку). Потом к данной паре userId|geo1 прикручиваем широту, также через ключ-склейку userId & "/" & geo1; в итоге получим табличку вида userId|geo1|geo2 без дублей по юзверям, которую можно прикрутить к логам визитов, также через left или outer. Правда, последний вариант предполагает, что получившаяся пара гео-координат - это место жительства/ место работы, иными словами, единственное место, из которого пользователь заказывает товар, что, мягко говоря, может не соответствовать суровой действительности.
